```python
from arcgis.gis import GIS
gis = GIS("home")
print("Logged in as " + gis.properties.user.username + " to " + gis.properties.name)
```

```python
help(gis.content.search)
```

```python
searchContent = gis.content.search("Occupation", item_type="Feature Layer", max_items=5, outside_org="True")
searchContent
```

```python
from IPython.display import display
for item in searchContent:
    display(item)
```

```python
item.title
```

```python
displayItem = searchContent[0]
display(displayItem)
```

```python
#Create a new map object
myMap = gis.map()
#Focus the map to the part of the world containing the example item
myMap.extent = displayItem.extent
#Display the map
myMap
```

```python
myMap.add_layer(displayItem)
```
